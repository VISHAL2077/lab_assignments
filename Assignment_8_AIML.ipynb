```python
car_train = train_data.drop('binaryClass', axis=1)
car_labels = train_data[['binaryClass']]
```

```python
from sklearn.preprocessing import OrdinalEncoder

from sklearn.pipeline import Pipeline

pipe = Pipeline([('ord_enc', OrdinalEncoder())])

car_train_prepared = pipe.fit_transform(car_train)
```

```python
from sklearn.preprocessing import LabelEncoder

label_enc = LabelEncoder()

car_labels_prepared = label_enc.fit_transform(car_labels)
```

```python
car_train_prepared.shape, car_labels_prepared.shape
```

```python
from sklearn.tree import DecisionTreeClassifier

tree_clf = DecisionTreeClassifier()
tree_clf.set_params(class_weight='balanced', criterion='entropy')

tree_clf.fit(car_train_prepared, car_labels_prepared)
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(20,15))
plot_tree(tree_clf, feature_names=car_train.columns,
          class_names=['N','P']);
```

```python

from sklearn.metrics import accuracy_score, classification_report,\
confusion_matrix

preds = tree_clf.predict(car_train_prepared)
print(f"Accuracy:{accuracy_score(car_labels_prepared, preds)}")
print(f"ClassReport:{classification_report(car_labels_prepared,preds)}")
print(f"Confusion Matrix:{confusion_matrix(car_labels_prepared,preds)}")
```

```python
from sklearn.model_selection import RandomizedSearchCV

params_dist = {
    'max_leaf_nodes': list(range(1, 10)),  # Avoid 0 as it doesn't make sense
    'min_samples_split': [1, 2, 3, 4],
    'min_samples_leaf': [1, 2, 3, 4],
    'max_depth': [1, 2, 3, 4, 5],
    'max_features': [1, 2, 3, 4, 5]
}

random_search = RandomizedSearchCV(DecisionTreeClassifier(random_state=42, class_weight='balanced'),
                                   params_dist,
                                   n_iter=100,  # Number of parameter settings to sample
                                   verbose=1,
                                   cv=3,
                                   refit=True,
                                   random_state=42)

random_search.fit(car_train_prepared, car_labels_prepared)
```

```python
try:
    best_params = random_search.best_params_
    print("Best Parameters:", best_params)
except AttributeError:
    print("The fitting process may not have completed successfully. Please check for errors.")
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV

params_dist = {
    'max_leaf_nodes': list(range(1, 10)),
    'min_samples_split': [1, 2, 3, 4],
    'min_samples_leaf': [1, 2, 3, 4],
    'max_depth': [1, 2, 3, 4, 5],
    'max_features': [1, 2, 3, 4, 5]
}

random_search = RandomizedSearchCV(
    DecisionTreeClassifier(random_state=42, class_weight='balanced'),
    params_dist,
    n_iter=100,
    verbose=1,
    cv=3,
    refit=True,
    random_state=42
)

random_search.fit(car_train_prepared, car_labels_prepared)

try:
    tree_best = random_search.best_estimator_
    print("Best Estimator:", tree_best)
except AttributeError:
    print("The fitting process may not have completed successfully. Please check for errors.")
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(20,15))
plot_tree(tree_best, feature_names=car_train.columns, class_names=['N','P']);
```

```python
from sklearn.metrics import accuracy_score

def accuracy(X, model, y):
    predictions = model.predict(X)
    return accuracy_score(y, predictions)

print(accuracy(car_train_prepared, tree_best, car_labels_prepared))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Generate predictions
predictions = tree_best.predict(car_train_prepared)

# Compute confusion matrix
cm = confusion_matrix(car_labels_prepared, predictions)

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Class 0', 'Class 1'],
            yticklabels=['Class 0', 'Class 1'])
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.title('Confusion Matrix')
plt.show()
```
